```python
import pandas as pd

df= pd.read_csv("ai_student_impact_dataset.csv")

print(df.describe())

# check Perceived_AI_Dependency,Institutional_Policy,Anxiety_Level_During_Exams

print(df["Perceived_AI_Dependency"].value_counts())
print(df["Institutional_Policy"].value_counts())    
print(df["Anxiety_Level_During_Exams"].value_counts())  
df.corr(numeric_only=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.info()
```

```python
df["GPA_increment"]= df["Post_Semester_GPA"] - df["Pre_Semester_GPA"]

df.corr(numeric_only=True)
```

```python
df=df[["GPA_increment","Weekly_GenAI_Hours","Traditional_Study_Hours"]]
df["study_group"]=pd.cut(df["Traditional_Study_Hours"],[0,5,10,15,20,25,30,35,float("inf")])
sns.scatterplot(data=df,x="Weekly_GenAI_Hours",y="GPA_increment",hue="study_group",palette='rocket_r')

plt.plot()
```

```python
df.corr(numeric_only=True)
```

Si può osservare come il GPA increment dipenda principalmente dalle ore di studio tradizional, l'uso di GenAI invece e leggermente inversamente correlato alle ore di studio tradizionali.
Inoltre da grafo notiamo che con aumento di ore di GenAI aumenta il rischio di avre decremento maggiore di GPA

```python
df["AI_usage_groups"]= pd.cut(df["Weekly_GenAI_Hours"], bins=[-1, 0, 5, 10,15,20,25,30,35,40, float('inf')], labels=["0", "1-5", "6-10", "11-15", "16-20", "21-25", "26-30", "31-35", "36-40", "40+"])
plt.yticks(rotation=90)
sns.barplot(data=df,x="AI_usage_groups",y="GPA_increment")
plt.plot()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x="Weekly_GenAI_Hours",y="Traditional_Study_Hours",hue="GPA_increment",palette="rocket")

plt.plot()
```

Scuro -> decremento
Possiamo vedere che chi dedica molte ore ad AI Gen invece che allo studio tradizionale spesso ha decremento sostanziale

```python
df["trad_prop"]=df["Traditional_Study_Hours"]/(df["Weekly_GenAI_Hours"]+df["Traditional_Study_Hours"])

df["tot_study"]=df["Weekly_GenAI_Hours"]+df["Traditional_Study_Hours"]
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x="trad_prop",y ="tot_study",hue="GPA_increment", palette="rocket")

plt.plot()
```

si nota come la quantità di ore totali allocate nello studio abbia grande impatto sull'incremento, inoltre lo studio tradizionale è più efficace. Studiare poco , ma in modo tradizionale in genere porta almeno a stabilità  del GPA, mentre possiamo osservare che nella zona ad con studio totale oltre i 30, ma con studio tradizionale inferiore al 20% del totale si ha un'altra probabilità di decremento.
